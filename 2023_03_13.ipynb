### 아이디를 통해서 puuid를 얻어야 함 : summoner-v4
### puuid를 통해서 matchid 얻어야 함 :match-v5
### matchid를 통해서 match 정보나 timeline을 api를 통해 조회 가능

### 3개의 함수
#### puuid를 구하는 함수
#### match_id를 구하는 함수
#### match, timeline을 가져오는 함수
#### 자신의 닉네임(롤 아이디)를 작성하면 puuid => matchid => 첫번째 matchid만 사용해서 match, timeline을 가져오는 함수
#### hide on bush 닉네임

```python
# 아이디를 통해 puuid를 구하는 함수
def get_puuid(user_id) :
    global my_riot_keys
    if ' ' in user_id :
        user_id = user_id.replace(' ', '%20')
        print('')
    url = f'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/{user_id}?api_key={my_riot_keys}'
    res = requests.get(url).json()
    return res['puuid']
```

```python
# match, timeline을 가져오는 함수
def return_match_timeline(match_id) :
    global my_riot_keys
    match_url = f'https://asia.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={my_riot_keys}'
    timeline_url = f'https://asia.api.riotgames.com/lol/match/v5/matches/{match_id}/timeline?api_key={my_riot_keys}'
    res_match = requests.get(match_url).json()
    res_timeline = requests.get(timeline_url).json()
    dict_res = {'match' : res_match, 'timeline' : res_timeline}
    return dict_res # res1, res2 도 가능 python은 리턴값 여러개가 가능함.
```
