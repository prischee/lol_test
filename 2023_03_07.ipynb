```python
### 숫자, 계산할 숫자, 계산 부호를 넣어서 함수를 만들기
# 계산 부호 잘못 입력 시 잘못된 입력입니다 띄우기
```

```python
def calc(num1, str, num2) :
    if str == '+' : 
        return num1 + num2
    elif str == '-' : 
        return num1 - num2
    elif str == '*' :
        return num1 * num2
    elif str == '/' :
        return num1 / num2
    else :
        print("잘못된 입력입니다.")
        return false
```

```python
calc(10, '+', 12)
```
