- kill_point라는 컬럼을 만들어서 (kills + assists /같은팀의 총 킬수)를 입력 (killerId를 이용하는방법 or kills를 이용하는 방법 )

```python
df_create = []
lane_win = []
first_blood = []
kill_point = []
for i in range(len(tmp)) :
    if tmp.iloc[i].G15 > tmp.iloc[i].enemy_g15 :
        lane_win.append(1)
    else:
        lane_win.append(0)
    if int(tmp.iloc[i].killerId.split("|")[0]) == tmp.iloc[i].participantId :
        first_blood.append(1)
    else :
        first_blood.append(0)
    kill_cnt = 0
    assist_cnt = 0
    team_kill = 0
    for j in range(len(tmp.iloc[i].killerId.split("|"))) :
        if tmp.iloc[i].participantId == int(tmp.iloc[i].killerId.split("|")[j]) :
                kill_cnt = kill_cnt+1
        if tmp.iloc[i].teamId == 100 :
            if int(tmp.iloc[i].killerId.split("|")[j]) in [1, 2, 3, 4, 5] :
                team_kill = team_kill + 1
        else :
            if int(tmp.iloc[i].killerId.split("|")[j]) in [6, 7, 8, 9, 10] :
                team_kill = team_kill + 1
        for k in range(len(tmp.iloc[29].assistId.split("|")[3].split(','))) :
            #print(tmp.iloc[i].assistId.split("|")[j].split(','))
            if str(tmp.iloc[i].participantId) == tmp.iloc[i].assistId.split("|")[j].split(',')[k] :
                assist_cnt = assist_cnt+1
        try :
            point = (kill_cnt + assist_cnt) / team_kill
        except :
            point = 'perfect!'

    kill_point.append(point)

tmp['lane_win'] = lane_win
tmp['first_blood'] = first_blood
tmp['kill_point'] = kill_point
```
